```python
from bs4 import BeautifulSoup as bs
import requests as req
from pathlib import Path
from urllib.request import Request, urlopen, urlretrieve
import urllib.request
from urllib.parse import unquote
import os
import time
import random
import multiprocessing
import json
from colorama import Fore, Back, Style

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'}
```

```python
def makePath(path):
    try:
        os.makedirs(path)
    except OSError as error:
        print()
#         print('directory \"'+path+'\" can not be created')
#                 print(Fore.BLUE, 'path exists')
#                 print(Style.RESET_ALL)
    
```

```python
def saveVideo(url):
    opener = urllib.request.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36')]

    urllib.request.install_opener(opener)
    
    makePath(unquote(url.split('/',3)[3].rsplit('/',1)[0]))
    
    urllib.request.urlretrieve(url, unquote(url.split('/',3)[3]))
```

```python
def savePdf(url):
#     source = req.get(url, headers = headers)
    
    opener = urllib.request.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36')]

    urllib.request.install_opener(opener)
    
    makePath(unquote(url.split('/',3)[3].rsplit('/',1)[0]))
    
    urllib.request.urlretrieve(url, unquote(url.split('/',3)[3]))
```

```python
def getLinksFrom(soup,url):
    list_dirs = []
    
    for link__ in soup.find_all('a'):
        if link__.parent.name == 'pre':
            if ';O' not in link__['href']: 
                list_dirs.append(url+link__['href'])


    del(list_dirs[1::2])
    if len(list_dirs)>0:
        list_dirs.pop(0)
    return list_dirs
```
